```python
#define converter: lon lat to x y coords
def converter(lon2,lat2):
    lon1=0
    lat1=0
    dx = (lon2-lon1)*40000*math.cos((lat1+lat2)*math.pi/360)/360
    dy = (lat2-lat1)*40000/360
    return(np.array([dx,dy]))
```

```python
add_up=0
grid_size=np.linalg.norm(converter(0.025,0.025)-converter(0,0))
for j in range(len(df)):
    for i in range(len(df.transpose())):
        if df[i][j]:
            add_up+=1
area=add_up*grid_size
print(area,"km^2")
```

```python
array=np.array([])
for j in range(len(df)):
    for i in range(len(df.transpose())):
        if df[i][j]:
            array=np.append(array,df[i][j])
dist=max(array)
print(dist,"km")
```
